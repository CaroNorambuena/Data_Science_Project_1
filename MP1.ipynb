4. Filtrar los atributos

Después de analizar cada dataframe, se observa que el df4 es el dataset más consistente, debido a que el ID de los clientes esta completo, ordenado de menor a mayor, sin datos faltantes ni redundantes. A su vez, dado que los ID de los clientes están en orden, la información de los atributos "SEX", "EDUCATION", "MARRIAGE", "AGE", "default.payment.next.month" también esta ordenada. Por ende, se ocupará el df4 como una base para construir el dataframe final.

Con respecto a los demás dataframes, se observa que df1, df2 y df3 contienen los datos faltantes (NA) de los atributos en el df4. Además, se notó que el df2 se complementa con el df3 en un atributo en particular.

Dado lo anterior, para poder generar una base de datos consolidada se trabajará por atributo, en el cual se filtrarán y ordenarán los datos para generar por cada atributo un dataframe. Es importante recalcar que todos los dataframes generados tendrán un atributo en común: "ID", el cual será necesario para realizar el merge final.

4.1 Trabajar con atributo "ID"

```python
df_ID = pd.DataFrame(df4["ID"])
df_ID
```

4.2 Trabajar con el atributo "Limit Balance"

```python

df2_ID_limit_balance = pd.DataFrame(df2[["ID", "LIMIT_BAL"]]).drop([10, 25, 30, 33, 42, 43, 50, 51])
df2_ID_limit_balance
```

```python

df3_ID_limit_balance = pd.DataFrame(df3.iloc[[10, 25, 30, 33, 42, 43], [1, 2]])
df3_ID_limit_balance
```

```python
df_ID_limit_balance = pd.concat([df2_ID_limit_balance, df3_ID_limit_balance], axis=0).sort_values(by = "ID", axis = 0)
df_ID_limit_balance
```

4.3 Trabajar con el atributo "Sex"

```python
df4_ID_sex = pd.DataFrame(df4[["ID", "SEX"]]).drop([13, 18, 22, 24, 36, 49])
df4_ID_sex
```

```python
df1_ID_sex = pd.DataFrame(df1.iloc[[13, 18, 22, 24, 36, 49], [1, 3]])
df1_ID_sex
```

```python
df_ID_sex = pd.concat([df4_ID_sex, df1_ID_sex], axis=0).sort_values(by = "ID", axis = 0)
df_ID_sex
```

4.4 Trabajar con atributo "Education"

```python
df4_ID_education = pd.DataFrame(df4[["ID", "EDUCATION"]]).drop([3, 6, 26, 27, 37, 47])
df4_ID_education
```

```python
df3_ID_education = pd.DataFrame(df3.iloc[[3, 6, 26, 27, 37, 47], [1, 3]])
df3_ID_education
```

```python
df_ID_education = pd.concat([df4_ID_education, df3_ID_education], axis=0).sort_values(by = "ID", axis = 0)
df_ID_education
```

4.5 Trabajar con atriibuto "Marriage"

```python
df_ID_marriage = pd.DataFrame(df4[["ID", "MARRIAGE"]])
df_ID_marriage
```

4.6 Trabajar con atributo "Age"

```python
df4_ID_age = pd.DataFrame(df4[["ID", "AGE"]]).drop([10, 25, 30, 33, 42, 43])
df4_ID_age
```

```python
df2_ID_age = pd.DataFrame(df2.iloc[[30, 33, 42, 43, 50, 51], [1, 3]])
df2_ID_age
```

4.10 Trabajar con el atributo "default pay next month"

```python
df_default_payment_next_month = pd.DataFrame(df4[["ID", "default.payment.next.month"]])
df_default_payment_next_month
```

5. Crear nuevo Dataframe

```python
dfa = pd.merge(df_ID,df_ID_limit_balance)
dfb = pd.merge(dfa,df_ID_sex)
dfc = pd.merge(dfb,df_ID_education)
dfd = pd.merge(dfc,df_ID_marriage)
dfe = pd.merge(dfd,df_ID_age)
dfg = pd.merge(dfe,df_ID_pay)
dfh = pd.merge(dfg,df_ID_billatm)
dfi = pd.merge(dfh,df_ID_payatm)
dfj = pd.merge(dfi,df_default_payment_next_month)
```

- Nuevo dataframe (df_new) con todos los datos filtrados:

```python
df_new = dfj
df_new
```

6. Homogenizar el nuevo Dataframe 

6.1 Se realiza una copia del nuevo Dataframe

```python
df__ = df_new.copy()
```

6.2 Trabajar con el atributo "Sex"

```python
df__['SEX'] = df__['SEX'].replace('1','male')
df__['SEX'] = df__['SEX'].replace('2','female')
```

6.3 Trabajar con el atributo "Education"

```python
df__['EDUCATION'] = df__['EDUCATION'].astype(float).astype(str).str[:-2]
```

6.4 Trabajar con el atributo "Age"

```python
df__['AGE'] = df__['AGE'].astype(str).str.replace('.0', ' years', regex=False)
```

7. Normalización del nuevo Dataframe

7.1 Antes de normalizar, se busca conocer los tipos de datos que contienen cada atributo del nuevo Dataframe. Solo "SEX" Y "AGE" serán de tipo object.

```python
df__.dtypes
```

7.2 Se transforman los datos de tipo objeto

Según el formato requerido...

- El Atributo "Limit balance" se transforma a float:

```python
df__['LIMIT_BAL'] = df__['LIMIT_BAL'].astype(float)
```

- El atributo "EDUCATION" se transforma en int:

```python
df__['EDUCATION'] = df__['EDUCATION'].astype(int)
```

7.3 Se transforman los datos de tipo int

Según el formato requerido, los atributos "BILL_AMT1", "BILL_AMT2", "BILL_AMT3", "BILL_AMT4", "BILL_AMT5", "BILL_AMT6" se transforman a float. Para ello, se utiliza FunctionTransformer de la librería scikit-learn que devuelve el resultado del la función trunc de la librería Numpy. De esta manera, los datos quedan en formato decimal (cero a la derecha).

```python
scaler = FunctionTransformer(np.trunc)

df__[["BILL_AMT1", "BILL_AMT2", "BILL_AMT3", \
 "BILL_AMT4", "BILL_AMT5", "BILL_AMT6"]] = scaler.fit_transform(df__[["BILL_AMT1", "BILL_AMT2", "BILL_AMT3", "BILL_AMT4", "BILL_AMT5", "BILL_AMT6"]])
```
